```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
#Get the data!
data = np.load('data/temp_data.npy')
print np.shape(data)
```

```python
#Visualize data
plt.scatter(data[:,0], data[:,1], c=data[:,2])
plt.show()

pos = data[data[:,2]==0.0]
plt.scatter(pos[:,0], pos[:,1])
plt.gca().set_xlim([0,10])
plt.gca().set_ylim([0,10])
plt.show()

neg = data[data[:,2]==1.0]
plt.scatter(neg[:,0], neg[:,1])
plt.gca().set_xlim([0,10])
plt.gca().set_ylim([0,10])
plt.show()
```

```python
#Initializing training set
X = data[:, 0:2]
Y = data[:, 2:].flatten()
print np.shape(X)
print np.shape(Y)

#Initialize svm model
clf = svm.SVC()
clf.fit(X, Y)
```
